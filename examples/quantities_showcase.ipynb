# evtol-tools Quantities System Showcase

This notebook demonstrates the complete capabilities of the evtol-tools units system, including:

- **19 Physical Quantity Types** - SI base and derived quantities
- **Unit Conversions** - Seamless conversion between compatible units
- **Arithmetic Operations** - Addition, subtraction, multiplication, division
- **Comparison Operations** - Full comparison support across units
- **NumPy Array Support** - Batch calculations with arrays
- **Cross-Type Arithmetic** - Derived quantities through multiplication/division
- **Flexible Unit Notation** - Any dimensionally-equivalent unit string accepted

```python
# Import all quantities and helper functions
from evtoltools.common import (
    # SI Base Quantities
    Mass, Length, Time, Temperature, Current, Substance, Luminosity,
    # Derived Quantities
    Velocity, Area, Volume, Force, Power, Density, Moment,
    AngularVelocity, Voltage, Energy, Capacity, Pressure, Frequency,
    # Helper functions
    in_units_of,
    # Advanced: direct pint access
    ureg, Q_
)
import numpy as np

print("All quantities imported successfully!")
```

## Part 1: SI Base Quantities

The seven SI base quantities form the foundation of the units system.

### 1.1 Mass

**Dimensionality:** `[mass]`  
**Default Unit:** kg  
**Common Units:** kg, g, lb, lbs, ton, tonne, oz

```python
# Mass - fundamental for aircraft weight calculations
aircraft_mass = Mass(1500, 'kg')
payload = Mass(200, 'lbs')

print(f"Aircraft mass: {aircraft_mass}")
print(f"Payload: {payload}")
print(f"Payload in kg: {payload.to('kg')}")
print(f"Payload value in kg: {payload.in_units_of('kg'):.2f} kg")

# Addition works across units
total_mass = aircraft_mass + payload
print(f"\nTotal mass: {total_mass}")
```

### 1.2 Length

**Dimensionality:** `[length]`  
**Default Unit:** m  
**Common Units:** m, cm, mm, ft, in, mi, km, nmi (nautical miles)

```python
# Length - critical for rotor diameter, wingspan, altitude
rotor_diameter = Length(2.0, 'm')
altitude = Length(1000, 'ft')
range_distance = Length(50, 'nmi')

print(f"Rotor diameter: {rotor_diameter}")
print(f"Altitude: {altitude} = {altitude.in_units_of('m'):.1f} m")
print(f"Range: {range_distance} = {range_distance.in_units_of('km'):.1f} km")
```

### 1.3 Time

**Dimensionality:** `[time]`  
**Default Unit:** s  
**Common Units:** s, ms, min, hr, day

```python
# Time - used for endurance, mission duration
hover_time = Time(15, 'min')
cruise_duration = Time(1.5, 'hr')
response_time = Time(500, 'ms')

print(f"Hover time: {hover_time} = {hover_time.in_units_of('s'):.0f} seconds")
print(f"Cruise duration: {cruise_duration} = {cruise_duration.in_units_of('min'):.0f} minutes")
print(f"Response time: {response_time}")

# Total mission time
total_time = hover_time + cruise_duration
print(f"\nTotal mission time: {total_time.to('min')}")
```

### 1.4 Temperature

**Dimensionality:** `[temperature]`  
**Default Unit:** K  
**Common Units:** K, degC, degF

```python
# Temperature - important for battery and motor thermal management
ambient_temp = Temperature(25, 'degC')
battery_max_temp = Temperature(140, 'degF')

print(f"Ambient temperature: {ambient_temp}")
print(f"Ambient in Kelvin: {ambient_temp.to('K')}")
print(f"Ambient in Fahrenheit: {ambient_temp.in_units_of('degF'):.1f} degF")
print(f"\nBattery max temp: {battery_max_temp} = {battery_max_temp.in_units_of('degC'):.1f} degC")
```

### 1.5 Electric Current

**Dimensionality:** `[current]`  
**Default Unit:** A  
**Common Units:** A, mA, uA, kA

```python
# Current - essential for electrical system sizing
motor_current = Current(150, 'A')
sensor_current = Current(50, 'mA')

print(f"Motor current: {motor_current}")
print(f"Sensor current: {sensor_current} = {sensor_current.in_units_of('A'):.3f} A")
```

### 1.6 Amount of Substance

**Dimensionality:** `[substance]`  
**Default Unit:** mol  
**Common Units:** mol, mmol, kmol

```python
# Substance - used in battery chemistry calculations
lithium_moles = Substance(2.5, 'mol')
electrolyte = Substance(500, 'mmol')

print(f"Lithium amount: {lithium_moles}")
print(f"Electrolyte: {electrolyte} = {electrolyte.in_units_of('mol'):.2f} mol")
```

### 1.7 Luminous Intensity

**Dimensionality:** `[luminosity]`  
**Default Unit:** cd  
**Common Units:** cd, lm

```python
# Luminosity - for navigation lights and displays
nav_light = Luminosity(100, 'cd')

print(f"Navigation light intensity: {nav_light}")
```

## Part 2: Derived Quantities

These quantities are derived from combinations of base quantities.

### 2.1 Velocity

**Dimensionality:** `[length] / [time]`  
**Default Unit:** m/s  
**Common Units:** m/s, km/h, mph, knots, ft/s

```python
# Velocity - cruise speed, climb rate
cruise_speed = Velocity(150, 'km/h')
climb_rate = Velocity(1000, 'ft/min')  # Equivalent unit string!
approach_speed = Velocity(60, 'knots')

print(f"Cruise speed: {cruise_speed}")
print(f"Cruise speed: {cruise_speed.in_units_of('mph'):.1f} mph")
print(f"Cruise speed: {cruise_speed.in_units_of('knots'):.1f} knots")
print(f"\nClimb rate: {climb_rate}")
print(f"Approach speed: {approach_speed} = {approach_speed.in_units_of('km/h'):.1f} km/h")
```

### 2.2 Area

**Dimensionality:** `[length] ** 2`  
**Default Unit:** m^2  
**Common Units:** m^2, cm^2, ft^2, in^2

```python
# Area - rotor disk area, wing area
disk_area = Area(3.14, 'm^2')
wing_area = Area(150, 'ft^2')

print(f"Rotor disk area: {disk_area}")
print(f"Wing area: {wing_area} = {wing_area.in_units_of('m^2'):.2f} m^2")
```

### 2.3 Volume

**Dimensionality:** `[length] ** 3`  
**Default Unit:** m^3  
**Common Units:** m^3, L, gal, ft^3

```python
# Volume - fuel tanks (for hybrid), cargo space
cargo_volume = Volume(500, 'L')
fuel_capacity = Volume(20, 'gal')

print(f"Cargo volume: {cargo_volume} = {cargo_volume.in_units_of('m^3'):.3f} m^3")
print(f"Fuel capacity: {fuel_capacity} = {fuel_capacity.in_units_of('L'):.1f} L")
```

### 2.4 Force

**Dimensionality:** `[mass] * [length] / [time] ** 2`  
**Default Unit:** N  
**Common Units:** N, kN, lbf, kgf

```python
# Force - thrust, weight, drag
thrust = Force(5000, 'N')
weight = Force(3000, 'lbf')

print(f"Thrust: {thrust} = {thrust.in_units_of('kN'):.2f} kN")
print(f"Weight: {weight} = {weight.in_units_of('N'):.1f} N")

# Equivalent unit notation: Force can be created with base units
force_from_base = Force(10, 'kg*m/s^2')  # Equivalent to 10 N
print(f"\nForce from base units: {force_from_base.in_units_of('N'):.1f} N")
```

### 2.5 Power

**Dimensionality:** `[mass] * [length] ** 2 / [time] ** 3`  
**Default Unit:** W  
**Common Units:** W, kW, MW, hp, bhp

```python
# Power - motor power, hover power
motor_power = Power(50, 'kW')
hover_power = Power(150, 'hp')

print(f"Motor power: {motor_power} = {motor_power.in_units_of('hp'):.1f} hp")
print(f"Hover power: {hover_power} = {hover_power.in_units_of('kW'):.1f} kW")

# Total power for 4-motor configuration
total_power = motor_power * 4
print(f"\nTotal power (4 motors): {total_power}")
```

### 2.6 Density

**Dimensionality:** `[mass] / [length] ** 3`  
**Default Unit:** kg/m^3  
**Common Units:** kg/m^3, g/cm^3, lb/ft^3

```python
# Density - air density, material density
air_density = Density(1.225, 'kg/m^3')  # Sea level ISA
aluminum_density = Density(2.7, 'g/cm^3')

print(f"Air density (sea level): {air_density}")
print(f"Aluminum density: {aluminum_density} = {aluminum_density.in_units_of('kg/m^3'):.0f} kg/m^3")
```

### 2.7 Moment (Torque)

**Dimensionality:** `[mass] * [length] ** 2 / [time] ** 2`  
**Default Unit:** N*m  
**Common Units:** N*m, ft*lbf, in*lbf

```python
# Moment - motor torque, aerodynamic moments
motor_torque = Moment(100, 'N*m')
control_moment = Moment(50, 'ft*lbf')

print(f"Motor torque: {motor_torque}")
print(f"Control moment: {control_moment} = {control_moment.in_units_of('N*m'):.1f} N*m")
```

### 2.8 Angular Velocity

**Dimensionality:** `1 / [time]` (with radian as dimensionless)  
**Default Unit:** rad/s  
**Common Units:** rad/s, rpm, deg/s

```python
# Angular velocity - rotor speed, motor speed
rotor_speed = AngularVelocity(2000, 'rpm')
motor_speed = AngularVelocity(5000, 'rpm')

print(f"Rotor speed: {rotor_speed}")
print(f"Rotor speed: {rotor_speed.in_units_of('rad/s'):.1f} rad/s")
print(f"Motor speed: {motor_speed} = {motor_speed.in_units_of('rad/s'):.1f} rad/s")
```

### 2.9 Voltage

**Dimensionality:** `[mass] * [length] ** 2 / [current] / [time] ** 3`  
**Default Unit:** V  
**Common Units:** V, mV, kV

```python
# Voltage - battery voltage, motor voltage
battery_voltage = Voltage(400, 'V')
cell_voltage = Voltage(3700, 'mV')

print(f"Battery pack voltage: {battery_voltage}")
print(f"Cell voltage: {cell_voltage} = {cell_voltage.in_units_of('V'):.2f} V")
```

### 2.10 Energy

**Dimensionality:** `[mass] * [length] ** 2 / [time] ** 2`  
**Default Unit:** J  
**Common Units:** J, kJ, MJ, Wh, kWh, mWh

```python
# Energy - battery capacity, mission energy
battery_energy = Energy(50, 'kWh')
mission_energy = Energy(30, 'kWh')

print(f"Battery energy: {battery_energy}")
print(f"Battery energy: {battery_energy.in_units_of('MJ'):.1f} MJ")
print(f"Mission energy: {mission_energy}")

# Remaining energy
remaining = battery_energy - mission_energy
print(f"\nRemaining energy: {remaining}")
```

### 2.11 Capacity (Electric Charge)

**Dimensionality:** `[current] * [time]`  
**Default Unit:** Ah  
**Common Units:** Ah, mAh, C (Coulombs)

```python
# Capacity - battery cell capacity
cell_capacity = Capacity(5000, 'mAh')
pack_capacity = Capacity(100, 'Ah')

print(f"Cell capacity: {cell_capacity} = {cell_capacity.in_units_of('Ah'):.1f} Ah")
print(f"Pack capacity: {pack_capacity}")
print(f"Pack capacity: {pack_capacity.in_units_of('C'):.0f} Coulombs")
```

### 2.12 Pressure

**Dimensionality:** `[mass] / [length] / [time] ** 2`  
**Default Unit:** Pa  
**Common Units:** Pa, kPa, bar, psi, atm

```python
# Pressure - atmospheric pressure, disk loading
atmospheric = Pressure(101325, 'Pa')
disk_loading = Pressure(50, 'lbf/ft^2')  # Equivalent unit string!
tire_pressure = Pressure(35, 'psi')

print(f"Atmospheric pressure: {atmospheric}")
print(f"Atmospheric pressure: {atmospheric.in_units_of('atm'):.3f} atm")
print(f"Atmospheric pressure: {atmospheric.in_units_of('bar'):.3f} bar")
print(f"\nDisk loading: {disk_loading} = {disk_loading.in_units_of('Pa'):.1f} Pa")
print(f"Tire pressure: {tire_pressure} = {tire_pressure.in_units_of('kPa'):.1f} kPa")

# Equivalent unit notation: Pressure can be created with N/m^2
pressure_from_base = Pressure(1000, 'N/m^2')  # Equivalent to 1000 Pa
print(f"\nPressure from N/m^2: {pressure_from_base.in_units_of('kPa'):.1f} kPa")
```

### 2.13 Frequency

**Dimensionality:** `1 / [time]`  
**Default Unit:** Hz  
**Common Units:** Hz, kHz, MHz, mHz

```python
# Frequency - control loop rate, sensor sampling
control_freq = Frequency(400, 'Hz')
sensor_rate = Frequency(1, 'kHz')
radio_freq = Frequency(2.4, 'MHz')

print(f"Control loop frequency: {control_freq}")
print(f"Sensor sampling rate: {sensor_rate} = {sensor_rate.in_units_of('Hz'):.0f} Hz")
print(f"Radio frequency: {radio_freq}")
```

## Part 3: All Operations on Quantities

This section demonstrates every operation available on quantity objects.

### 3.1 Construction Methods

```python
# Method 1: Value and unit
m1 = Mass(100, 'kg')

# Method 2: Default unit (no unit specified)
m2 = Mass(200)  # Uses default 'kg'

# Method 3: From pint Quantity
pint_qty = Q_(150, 'kg')
m3 = Mass(pint_qty)

# Method 4: From cross-type arithmetic result
force_result = Force(100, 'N') * Length(2, 'm')  # Returns pint Quantity
m4 = Moment(force_result)  # Wrap in appropriate type

# Method 5: Equivalent unit notation
m5 = Force(10, 'kg*m/s^2')  # Same as Force(10, 'N')

print(f"From value+unit: {m1}")
print(f"From default unit: {m2}")
print(f"From pint Quantity: {m3}")
print(f"From cross-type result: {m4}")
print(f"From equivalent units: {m5} = {m5.in_units_of('N'):.0f} N")
```

### 3.2 Unit Conversion Methods

```python
mass = Mass(100, 'kg')

# .to(unit) - Returns new quantity instance in target units
mass_lbs = mass.to('lbs')
print(f"to('lbs'): {mass_lbs}")

# .to_default() - Returns new instance in default units
mass_from_lbs = Mass(220, 'lbs')
mass_default = mass_from_lbs.to_default()
print(f"to_default(): {mass_from_lbs} -> {mass_default}")

# .in_units_of(unit) - Returns numeric VALUE (not quantity)
value = mass.in_units_of('lbs')
print(f"in_units_of('lbs'): {value:.2f} (type: {type(value).__name__})")
```

### 3.3 Properties

```python
energy = Energy(50, 'kWh')

# .magnitude - Numeric value in current units
print(f"magnitude: {energy.magnitude}")

# .value - Alias for magnitude
print(f"value: {energy.value}")

# .units - Current unit as string
print(f"units: {energy.units}")
```

### 3.4 Arithmetic Operations

```python
m1 = Mass(100, 'kg')
m2 = Mass(50, 'kg')
m3 = Mass(30, 'lbs')  # Different units

# Addition - works across units, returns same type
sum_mass = m1 + m2
print(f"Addition (same units): {m1} + {m2} = {sum_mass}")

sum_mixed = m1 + m3
print(f"Addition (mixed units): {m1} + {m3} = {sum_mixed}")

# Subtraction
diff = m1 - m2
print(f"Subtraction: {m1} - {m2} = {diff}")

# Scalar multiplication - returns same type
doubled = m1 * 2
print(f"Scalar multiply: {m1} * 2 = {doubled}")

# Reverse scalar multiplication
tripled = 3 * m1
print(f"Reverse multiply: 3 * {m1} = {tripled}")

# Scalar division - returns same type
halved = m1 / 2
print(f"Scalar divide: {m1} / 2 = {halved}")
```

### 3.5 Comparison Operations

```python
m1 = Mass(100, 'kg')
m2 = Mass(220, 'lbs')  # Approximately 100 kg
m3 = Mass(50, 'kg')

# Equality (works across units)
print(f"{m1} == {m2}: {m1 == m2}")

# Less than
print(f"{m3} < {m1}: {m3 < m1}")

# Greater than
print(f"{m1} > {m3}: {m1 > m3}")

# Less than or equal
print(f"{m3} <= {m1}: {m3 <= m1}")

# Greater than or equal
print(f"{m1} >= {m3}: {m1 >= m3}")
```

### 3.6 Representation Methods

```python
power = Power(12345.678, 'W')

# str() - Human readable
print(f"str(): {str(power)}")

# repr() - Unambiguous, shows constructor form
print(f"repr(): {repr(power)}")

# Format specification
print(f"Format .2f: {power:.2f}")
print(f"Format .1e: {power:.1e}")
```

## Part 4: NumPy Array Support

All quantities support NumPy arrays for batch calculations.

```python
# Create quantity arrays
component_masses = Mass(np.array([50, 100, 150, 200]), 'kg')
print(f"Component masses: {component_masses}")
print(f"Magnitude type: {type(component_masses.magnitude)}")

# Array conversion
masses_lbs = component_masses.to('lbs')
print(f"\nIn pounds: {masses_lbs}")

# Get numeric array in specific units
values = component_masses.in_units_of('g')
print(f"Values in grams: {values}")
```

```python
# Array arithmetic
masses = Mass(np.array([100, 200, 300]), 'kg')
margin = Mass(10, 'kg')

# Broadcasting: scalar + array
with_margin = masses + margin
print(f"With margin: {with_margin}")

# Element-wise operations
multipliers = np.array([1.0, 1.1, 1.2])
scaled = masses * multipliers
print(f"Scaled: {scaled}")
```

```python
# Array comparisons return boolean arrays
masses = Mass(np.array([80, 120, 180, 250]), 'kg')
threshold = Mass(150, 'kg')

above_threshold = masses > threshold
print(f"Masses: {masses}")
print(f"Threshold: {threshold}")
print(f"Above threshold: {above_threshold}")

# Use with numpy functions
heavy_masses = masses.magnitude[above_threshold]
print(f"Heavy component masses: {heavy_masses} kg")
```

```python
# Aggregation with numpy
component_masses = Mass(np.array([120, 85, 43, 15, 200]), 'kg')

total = component_masses.magnitude.sum()
average = component_masses.magnitude.mean()
maximum = component_masses.magnitude.max()
minimum = component_masses.magnitude.min()

print(f"Components: {component_masses}")
print(f"Total: {total} kg")
print(f"Average: {average:.1f} kg")
print(f"Max: {maximum} kg")
print(f"Min: {minimum} kg")
```

## Part 5: Cross-Type Arithmetic and Derived Quantities

When multiplying or dividing different quantity types, the result is a pint Quantity. Use the `in_units_of()` helper function to extract values, or wrap the result in the appropriate type.

### 5.1 Fundamental Derived Quantities

```python
# Velocity = Length / Time
distance = Length(100, 'km')
time = Time(2, 'hr')
velocity_result = distance / time

print("VELOCITY = LENGTH / TIME")
print(f"{distance} / {time} = {velocity_result}")
print(f"Value in m/s: {in_units_of(velocity_result, 'm/s'):.2f}")
print(f"Value in km/h: {in_units_of(velocity_result, 'km/h'):.1f}")

# Wrap in Velocity type for further typed operations
v = Velocity(velocity_result)
print(f"As Velocity type: {v}")
```

```python
# Area = Length * Length
width = Length(10, 'm')
height = Length(5, 'm')
area_result = width * height

print("AREA = LENGTH * LENGTH")
print(f"{width} * {height} = {area_result}")
print(f"Value in m^2: {in_units_of(area_result, 'm^2'):.1f}")
print(f"Value in ft^2: {in_units_of(area_result, 'ft^2'):.1f}")

# Wrap in Area type
a = Area(area_result)
print(f"As Area type: {a}")
```

```python
# Volume = Length * Length * Length (or Area * Length)
length = Length(2, 'm')
width = Length(3, 'm')
height = Length(4, 'm')

volume_result = length * width * height

print("VOLUME = LENGTH * LENGTH * LENGTH")
print(f"{length} * {width} * {height} = {volume_result}")
print(f"Value in m^3: {in_units_of(volume_result, 'm^3'):.1f}")
print(f"Value in L: {in_units_of(volume_result, 'L'):.0f}")

v = Volume(volume_result)
print(f"As Volume type: {v}")
```

```python
# Force = Mass * Acceleration (Length / Time^2)
mass = Mass(10, 'kg')
acceleration = Length(9.81, 'm') / (Time(1, 's') * Time(1, 's'))

force_result = mass * acceleration

print("FORCE = MASS * ACCELERATION")
print(f"{mass} * {acceleration} = {force_result}")
print(f"Value in N: {in_units_of(force_result, 'N'):.2f}")

f = Force(force_result)
print(f"As Force type: {f}")
```

### 5.2 Mechanical Derived Quantities

```python
# Moment (Torque) = Force * Length
force = Force(100, 'N')
arm = Length(0.5, 'm')
moment_result = force * arm

print("MOMENT = FORCE * LENGTH")
print(f"{force} * {arm} = {moment_result}")
print(f"Value in N*m: {in_units_of(moment_result, 'N*m'):.1f}")
print(f"Value in ft*lbf: {in_units_of(moment_result, 'ft*lbf'):.2f}")

# Verify equality with Moment type
m = Moment(moment_result)
m_direct = Moment(50, 'N*m')
print(f"\nMoment(result) == Moment(50, 'N*m'): {m == m_direct}")
```

```python
# Power = Force * Velocity (or Energy / Time)
thrust = Force(5000, 'N')
speed = Velocity(50, 'm/s')
power_result = thrust * speed

print("POWER = FORCE * VELOCITY")
print(f"{thrust} * {speed} = {power_result}")
print(f"Value in W: {in_units_of(power_result, 'W'):.0f}")
print(f"Value in kW: {in_units_of(power_result, 'kW'):.0f}")
print(f"Value in hp: {in_units_of(power_result, 'hp'):.1f}")

p = Power(power_result)
print(f"As Power type: {p}")
```

```python
# Energy = Power * Time
power = Power(50, 'kW')
duration = Time(30, 'min')
energy_result = power * duration

print("ENERGY = POWER * TIME")
print(f"{power} * {duration} = {energy_result}")
print(f"Value in J: {in_units_of(energy_result, 'J'):.0f}")
print(f"Value in kWh: {in_units_of(energy_result, 'kWh'):.1f}")

e = Energy(energy_result)
print(f"As Energy type: {e}")
```

```python
# Pressure = Force / Area
force = Force(1000, 'N')
area = Area(0.1, 'm^2')
pressure_result = force / area

print("PRESSURE = FORCE / AREA")
print(f"{force} / {area} = {pressure_result}")
print(f"Value in Pa: {in_units_of(pressure_result, 'Pa'):.0f}")
print(f"Value in kPa: {in_units_of(pressure_result, 'kPa'):.1f}")
print(f"Value in psi: {in_units_of(pressure_result, 'psi'):.2f}")

p = Pressure(pressure_result)
print(f"As Pressure type: {p}")
```

```python
# Density = Mass / Volume
mass = Mass(2700, 'kg')
volume = Volume(1, 'm^3')
density_result = mass / volume

print("DENSITY = MASS / VOLUME")
print(f"{mass} / {volume} = {density_result}")
print(f"Value in kg/m^3: {in_units_of(density_result, 'kg/m^3'):.0f}")
print(f"Value in g/cm^3: {in_units_of(density_result, 'g/cm^3'):.2f}")

d = Density(density_result)
print(f"As Density type: {d}")
```

### 5.3 Electrical Derived Quantities

```python
# Electrical Power = Voltage * Current
voltage = Voltage(400, 'V')
current = Current(100, 'A')
electrical_power = voltage * current

print("ELECTRICAL POWER = VOLTAGE * CURRENT")
print(f"{voltage} * {current} = {electrical_power}")
print(f"Value in W: {in_units_of(electrical_power, 'W'):.0f}")
print(f"Value in kW: {in_units_of(electrical_power, 'kW'):.0f}")

p = Power(electrical_power)
print(f"As Power type: {p}")
```

```python
# Capacity = Current * Time
current = Current(5, 'A')
time = Time(1, 'hr')
capacity_result = current * time

print("CAPACITY = CURRENT * TIME")
print(f"{current} * {time} = {capacity_result}")
print(f"Value in Ah: {in_units_of(capacity_result, 'Ah'):.1f}")
print(f"Value in C: {in_units_of(capacity_result, 'C'):.0f}")

c = Capacity(capacity_result)
print(f"As Capacity type: {c}")
```

### 5.4 Dimensionless Ratios

## Part 6: Practical eVTOL Calculations

Complete examples demonstrating real-world eVTOL engineering calculations.

### 6.1 Hover Power Calculation

```python
# Hover Power using Momentum Theory
# P_ideal = T * sqrt(T / (2 * rho * A))

# Vehicle parameters
mtow = Mass(500, 'kg')
num_rotors = 4
rotor_diameter = Length(2, 'm')
air_density = Density(1.225, 'kg/m^3')
figure_of_merit = 0.7  # Efficiency factor
motor_efficiency = 0.9

# Calculate thrust (weight)
g = 9.81  # m/s^2
thrust = Force(mtow.magnitude * g, 'N')
print(f"Vehicle weight: {thrust}")

# Calculate total disk area
rotor_radius = rotor_diameter / 2
single_disk_area = Area(np.pi * rotor_radius.magnitude**2, 'm^2')
total_disk_area = single_disk_area * num_rotors
print(f"Single rotor disk area: {single_disk_area}")
print(f"Total disk area: {total_disk_area}")

# Disk loading
disk_loading = thrust / total_disk_area
print(f"Disk loading: {in_units_of(disk_loading, 'N/m^2'):.1f} N/m^2")

# Ideal hover power: P = T * sqrt(T / (2 * rho * A))
T = thrust.magnitude  # N
rho = air_density.magnitude  # kg/m^3
A = total_disk_area.magnitude  # m^2
P_ideal = T * np.sqrt(T / (2 * rho * A))
ideal_power = Power(P_ideal, 'W')
print(f"\nIdeal hover power: {ideal_power.to('kW')}")

# Actual shaft power (accounting for rotor efficiency)
shaft_power = Power(P_ideal / figure_of_merit, 'W')
print(f"Shaft power (FM={figure_of_merit}): {shaft_power.to('kW')}")

# Electrical power (accounting for motor efficiency)
electrical_power = Power(shaft_power.magnitude / motor_efficiency, 'W')
print(f"Electrical power (eta={motor_efficiency}): {electrical_power.to('kW')}")
```

### 6.2 Battery Sizing

```python
# Battery sizing for hover endurance

# Requirements
hover_power = Power(45, 'kW')
target_endurance = Time(20, 'min')
usable_capacity_fraction = 0.8  # Only use 80% of battery capacity

# Cell specifications
cell_voltage = Voltage(3.7, 'V')
cell_capacity = Capacity(5000, 'mAh')
cell_mass = Mass(70, 'g')
pack_voltage = Voltage(48, 'V')

# Calculate required energy
required_energy_result = hover_power * target_endurance
required_energy = Energy(required_energy_result)
print(f"Required hover energy: {required_energy.to('Wh')}")

# Total battery energy needed (accounting for usable fraction)
total_energy = Energy(required_energy.magnitude / usable_capacity_fraction, 'Wh')
print(f"Total battery energy (at {usable_capacity_fraction*100:.0f}% usable): {total_energy}")

# Calculate cell configuration
cells_in_series = int(np.ceil(pack_voltage.magnitude / cell_voltage.magnitude))
print(f"\nCells in series: {cells_in_series}")

# Energy per string
cell_energy = Energy(in_units_of(cell_capacity * cell_voltage, 'Wh'), 'Wh')
string_energy = Energy(cell_energy.magnitude * cells_in_series, 'Wh')
print(f"Cell energy: {cell_energy}")
print(f"String energy: {string_energy}")

# Parallel strings needed
parallel_strings = int(np.ceil(total_energy.magnitude / string_energy.magnitude))
print(f"Parallel strings needed: {parallel_strings}")

# Total cells and mass
total_cells = cells_in_series * parallel_strings
battery_mass = Mass(total_cells * cell_mass.magnitude, 'g')
print(f"\nTotal cells: {total_cells}")
print(f"Battery mass: {battery_mass.to('kg')}")

# Actual battery capacity
actual_energy = Energy(string_energy.magnitude * parallel_strings, 'Wh')
print(f"Actual battery energy: {actual_energy}")

# Specific energy
specific_energy = actual_energy / battery_mass
print(f"Specific energy: {in_units_of(specific_energy, 'Wh/kg'):.1f} Wh/kg")
```

### 6.3 Mission Profile Analysis

```python
# Simple mission profile energy calculation

# Mission phases with power and duration (using arrays!)
phases = ['Takeoff/Climb', 'Cruise', 'Descent', 'Landing']
powers = Power(np.array([60, 35, 20, 50]), 'kW')
durations = Time(np.array([2, 15, 3, 1]), 'min')

print("MISSION PROFILE ANALYSIS")
print("=" * 50)

# Calculate energy for each phase
energies_result = powers._quantity * durations._quantity
energies_wh = energies_result.to('Wh').magnitude

for i, phase in enumerate(phases):
    print(f"{phase:15s}: {powers.magnitude[i]:5.0f} kW x {durations.magnitude[i]:4.0f} min = {energies_wh[i]:7.1f} Wh")

print("-" * 50)
total_energy = Energy(energies_wh.sum(), 'Wh')
total_time = Time(durations.magnitude.sum(), 'min')
print(f"{'TOTAL':15s}: {total_energy} over {total_time}")

# Average power
avg_power = total_energy / total_time
print(f"Average power: {in_units_of(avg_power, 'kW'):.1f} kW")
```

## Part 7: Equivalent Unit Strings

The units system accepts any dimensionally-equivalent unit string, not just the common units listed in the documentation.

```python
# Pressure can be expressed in multiple equivalent ways
p1 = Pressure(1000, 'Pa')
p2 = Pressure(1000, 'N/m^2')       # Equivalent to Pa
p3 = Pressure(1000, 'kg/(m*s^2)')  # Also equivalent

print("Pressure equivalents:")
print(f"  1000 Pa      = {p1}")
print(f"  1000 N/m^2   = {p2}")
print(f"  1000 kg/(m*s^2) = {p3}")
print(f"\nAll equal to Pa: {p1 == p2 == p3}")
```

```python
# Force equivalents
f1 = Force(10, 'N')
f2 = Force(10, 'kg*m/s^2')  # Newton's definition

print("Force equivalents:")
print(f"  10 N        = {f1}")
print(f"  10 kg*m/s^2 = {f2}")
print(f"\nBoth equal: {f1 == f2}")
```

```python
# Velocity equivalents
v1 = Velocity(10, 'm/s')
v2 = Velocity(10, 'm*s^-1')  # Using negative exponent

print("Velocity equivalents:")
print(f"  10 m/s     = {v1}")
print(f"  10 m*s^-1  = {v2}")
print(f"\nBoth equal: {v1 == v2}")
```

```python
# Conversion to equivalent units also works
pressure = Pressure(101325, 'Pa')

print(f"Atmospheric pressure: {pressure}")
print(f"  in N/m^2: {pressure.in_units_of('N/m^2'):.0f}")
print(f"  in kg/(m*s^2): {pressure.in_units_of('kg/(m*s^2)'):.0f}")
print(f"  in atm: {pressure.in_units_of('atm'):.4f}")
```
